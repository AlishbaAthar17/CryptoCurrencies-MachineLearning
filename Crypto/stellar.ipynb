```python
# making output column one day ahead of the closing price column
STELLAR['Target_Price'] = STELLAR['Close'].shift(-1)
STELLAR
```

```python
STELLAR.dropna(inplace=True)
```

```python
STELLAR.isnull().sum()
```

```python
# input features
X_STELLAR=STELLAR[['Open', 'Close']]

# output features
y_STELLAR=STELLAR['Target_Price']
```

# Training and Testing the Model

## We will give 0.10 % data to our model for testing to make accurate and close predictions

```python
X_STELLAR_train, X_STELLAR_test, y_STELLAR_train, y_STELLAR_test=train_test_split(X_STELLAR, y_STELLAR, test_size=.30, random_state=50)
```

```python
model_STELLAR=LinearRegression()
model_STELLAR.fit(X_STELLAR_train, y_STELLAR_train)
```

```python
model_STELLAR.score(X_STELLAR_test, y_STELLAR_test)
```

```python
STELLAR_model=joblib.dump(model_STELLAR,"STELLAR_predictions.pkl")
```
