```python
# making output column one day ahead of the closing price column
LINK['Target_Price'] = LINK['Close'].shift(-1)
LINK
```

```python
LINK.isnull().sum()
```

```python
LINK.dropna(inplace=True)
```

```python
LINK.isnull().sum()
```

```python
# input features
X_LINK=LINK[['Open', 'Close']]

# output features
y_LINK=LINK['Target_Price']
```

# Training and Testing the Model

## We will give 0.10 % data to our model for testing to make accurate and close predictions

```python
X_LINK_train, X_LINK_test, y_LINK_train, y_LINK_test=train_test_split(X_LINK, y_LINK, test_size=.10, random_state=50)
```

```python
model_LINK=LinearRegression()
model_LINK.fit(X_LINK_train, y_LINK_train)
```

```python
model_LINK.score(X_LINK_test, y_LINK_test)
```

```python
LINK_model=joblib.dump(model_LINK,"LINK_predictions.pkl")
```
