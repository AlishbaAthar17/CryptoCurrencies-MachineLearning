```python
# making output column one day ahead of the closing price column
SHIBA['Target_Price'] = SHIBA['Close'].shift(-1)
SHIBA
```

```python
SHIBA.isnull().sum()
```

```python
SHIBA.dropna(inplace=True)
```

```python
SHIBA.isnull().sum()
```

```python
# input features
X_SHIBA=SHIBA[['Open', 'Close']]

# output features
y_SHIBA=SHIBA['Target_Price']
```

# Training and Testing the Model

## We will give 0.10 % data to our model for testing to make accurate and close predictions

```python
X_SHIBA_train, X_SHIBA_test, y_SHIBA_train, y_SHIBA_test=train_test_split(X_SHIBA, y_SHIBA, test_size=.30, random_state=50)
```

```python
model_SHIBA=LinearRegression()
model_SHIBA.fit(X_SHIBA_train, y_SHIBA_train)
```

```python
model_SHIBA.score(X_SHIBA_test, y_SHIBA_test)
```

```python
SHIBA_model=joblib.dump(model_SHIBA,"SHIBA_predictions.pkl")
```
