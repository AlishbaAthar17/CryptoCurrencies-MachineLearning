```python
# making output column one day ahead of the closing price column
POLYGON['Target_Price'] = POLYGON['Close'].shift(-1)
POLYGON
```

```python
POLYGON.isnull().sum()
```

```python
POLYGON.dropna(inplace=True)
```

```python
POLYGON.isnull().sum()
```

```python
# input features
X_POLYGON=POLYGON[['Open', 'Close']]

# output features
y_POLYGON=POLYGON['Target_Price']
```

# Training and Testing the Model

## We will give 0.10 % data to our model for testing to make accurate and close predictions

```python
X_POLYGON_train, X_POLYGON_test, y_POLYGON_train, y_POLYGON_test=train_test_split(X_POLYGON, y_POLYGON, test_size=.10, random_state=50)
```

```python
model_POLYGON=LinearRegression()
model_POLYGON.fit(X_POLYGON_train, y_POLYGON_train)
```

```python
model_POLYGON.score(X_POLYGON_test, y_POLYGON_test)
```

```python
POLYGON_model=joblib.dump(model_POLYGON,"POLYGON_predictions.pkl")
```
