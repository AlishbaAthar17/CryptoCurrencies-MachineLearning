```python
# making output column one day ahead of the closing price column
WBTC['Target_Price'] = WBTC['Close'].shift(-1)
WBTC
```

```python
WBTC.isnull().sum()
```

```python
WBTC.dropna(inplace=True)
```

```python
WBTC.isnull().sum()
```

```python
# input features
X_WBTC=WBTC[['Open', 'Close']]

# output features
y_WBTC=WBTC['Target_Price']
```

# Training and Testing the Model

## We will give 0.10 % data to our model for testing to make accurate and close predictions

```python
X_WBTC_train, X_WBTC_test, y_WBTC_train, y_WBTC_test=train_test_split(X_WBTC, y_WBTC, test_size=.10, random_state=50)
```

```python
model_WBTC=LinearRegression()
model_WBTC.fit(X_WBTC_train, y_WBTC_train)
```

```python
model_WBTC.score(X_WBTC_test, y_WBTC_test)
```

```python
WBTC_model=joblib.dump(model_WBTC,"WBTC_predictions.pkl")
```
