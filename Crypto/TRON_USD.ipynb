```python
# making output column one day ahead of the closing price column
TRON['Target_Price'] = TRON['Close'].shift(-1)
TRON
```

```python
TRON.isnull().sum()
```

```python
TRON.dropna(inplace=True)
```

```python
TRON.isnull().sum()
```

```python
# input features
X_TRON=TRON[['Open', 'Close']]

# output features
y_TRON=TRON['Target_Price']
```

# Training and Testing the Model

## We will give 0.10 % data to our model for testing to make accurate and close predictions

```python
X_TRON_train, X_TRON_test, y_TRON_train, y_TRON_test=train_test_split(X_TRON, y_TRON, test_size=.10, random_state=50)
```

```python
model_TRON=LinearRegression()
model_TRON.fit(X_TRON_train, y_TRON_train)
```

```python
model_TRON.score(X_TRON_test, y_TRON_test)
```

```python
TRON_model=joblib.dump(model_TRON,"TRON_predictions.pkl")
```
