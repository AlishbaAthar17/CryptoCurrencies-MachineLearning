```python
# making output column one day ahead of the closing price column
LITECOIN['Target_Price'] = LITECOIN['Close'].shift(-1)
LITECOIN
```

```python
LITECOIN.isnull().sum()
```

```python
LITECOIN.dropna(inplace=True)
```

```python
LITECOIN.isnull().sum()
```

```python
# input features
X_LITECOIN=LITECOIN[['Open', 'Close']]

# output features
y_LITECOIN=LITECOIN['Target_Price']
```

# Training and Testing the Model

## We will give 0.10 % data to our model for testing to make accurate and close predictions

```python
X_LITECOIN_train, X_LITECOIN_test, y_LITECOIN_train, y_LITECOIN_test=train_test_split(X_LITECOIN, y_LITECOIN, test_size=.10, random_state=50)
```

```python
model_LITECOIN=LinearRegression()
model_LITECOIN.fit(X_LITECOIN_train, y_LITECOIN_train)
```

```python
model_LITECOIN.score(X_LITECOIN_test, y_LITECOIN_test)
```

```python
LITECOIN_model=joblib.dump(model_LITECOIN,"LITECOIN_predictions.pkl")
```
