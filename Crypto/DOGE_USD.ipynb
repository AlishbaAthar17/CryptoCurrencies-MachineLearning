```python
# making output column one day ahead of the closing price column
DOGE['Target_Price'] = DOGE['Close'].shift(-1)
DOGE
```

```python
DOGE.isnull().sum()
```

```python
DOGE.dropna(inplace=True)
```

```python
DOGE.isnull().sum()
```

```python
# input features
X_DOGE=DOGE[['Open', 'Close']]

# output features
y_DOGE=DOGE['Target_Price']
```

# Training and Testing the Model

## We will give 0.10 % data to our model for testing to make accurate and close predictions

```python
X_DOGE_train, X_DOGE_test, y_DOGE_train, y_DOGE_test=train_test_split(X_DOGE, y_DOGE, test_size=.10, random_state=50)
```

```python
model_DOGE=LinearRegression()
model_DOGE.fit(X_DOGE_train, y_DOGE_train)
```

```python
model_DOGE.score(X_DOGE_test, y_DOGE_test)
```

```python
DOGE_model=joblib.dump(model_DOGE,"DOGE_predictions.pkl")
```
