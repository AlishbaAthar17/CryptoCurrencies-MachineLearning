```python
# making output column one day ahead of the closing price column
ADA['Target_Price'] = ADA['Close'].shift(-1)
ADA
```

```python
ADA.isnull().sum()
```

```python
ADA.dropna(inplace=True)
```

```python
ADA.isnull().sum()
```

```python
# input features
X_ADA=ADA[['Open', 'Close']]

# output features
y_ADA=ADA['Target_Price']
```

# Training and Testing the Model

## We will give 0.10 % data to our model for testing to make accurate and close predictions

```python
X_ADA_train, X_ADA_test, y_ADA_train, y_ADA_test=train_test_split(X_ADA, y_ADA, test_size=.10, random_state=50)
```

```python
model_ADA=LinearRegression()
model_ADA.fit(X_ADA_train, y_ADA_train)
```

```python
model_ADA.score(X_ADA_test, y_ADA_test)
```

```python
ADA_model=joblib.dump(model_ADA,"ADA_predictions.pkl")
```
