```python
# making output column one day ahead of the closing price column
DAI['Target_Price'] = DAI['Close'].shift(-1)
DAI
```

```python
DAI.isnull().sum()
```

```python
DAI.dropna(inplace=True)
```

```python
DAI.isnull().sum()
```

```python
# input features
X_DAI=DAI[['Open', 'Close']]

# output features
y_DAI=DAI['Target_Price']
```

# Training and Testing the Model

## We will give 0.10 % data to our model for testing to make accurate and close predictions

```python
X_DAI_train, X_DAI_test, y_DAI_train, y_DAI_test=train_test_split(X_DAI, y_DAI, test_size=.01, random_state=50)
```

```python
model_DAI=LinearRegression()
model_DAI.fit(X_DAI_train, y_DAI_train)
```

```python
model_DAI.score(X_DAI_test, y_DAI_test)
```

```python
DAI_model=joblib.dump(model_DAI,"DAI_predictions.pkl")
```
