```python
# making output column one day ahead of the closing price column
POLKADOT['Target_Price'] = POLKADOT['Close'].shift(-1)
POLKADOT
```

```python
POLKADOT.isnull().sum()
```

```python
POLKADOT.dropna(inplace=True)
```

```python
POLKADOT.isnull().sum()
```

```python
# input features
X_POLKADOT=POLKADOT[['Open', 'Close']]

# output features
y_POLKADOT=POLKADOT['Target_Price']
```

# Training and Testing the Model

## We will give 0.10 % data to our model for testing to make accurate and close predictions

```python
X_POLKADOT_train, X_POLKADOT_test, y_POLKADOT_train, y_POLKADOT_test=train_test_split(X_POLKADOT, y_POLKADOT, test_size=.10, random_state=50)
```

```python
model_POLKADOT=LinearRegression()
model_POLKADOT.fit(X_POLKADOT_train, y_POLKADOT_train)
```

```python
model_POLKADOT.score(X_POLKADOT_test, y_POLKADOT_test)
```

```python
POLKADOT_model=joblib.dump(model_POLKADOT,"POLKADOT_predictions.pkl")
```
