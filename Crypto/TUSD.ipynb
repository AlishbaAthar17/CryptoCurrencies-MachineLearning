```python
# making output column one day ahead of the closing price column
TrueUSD['Target_Price'] = TrueUSD['Close'].shift(-1)
TrueUSD
```

```python
TrueUSD.isnull().sum()
```

```python
TrueUSD.dropna(inplace=True)
```

```python
TrueUSD.isnull().sum()
```

```python
# input features
X_TrueUSD=TrueUSD[['Open', 'Close']]

# output features
y_TrueUSD=TrueUSD['Target_Price']
```

# Training and Testing the Model

## We will give 0.10 % data to our model for testing to make accurate and close predictions

```python
X_TrueUSD_train, X_TrueUSD_test, y_TrueUSD_train, y_TrueUSD_test=train_test_split(X_TrueUSD, y_TrueUSD, test_size=.01, random_state=50)
```

```python
model_TrueUSD=LinearRegression()
model_TrueUSD.fit(X_TrueUSD_train, y_TrueUSD_train)
```

```python
model_TrueUSD.score(X_TrueUSD_test, y_TrueUSD_test)
```

```python
TrueUSD_model=joblib.dump(model_TrueUSD,"TrueUSD_predictions.pkl")
```
